```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
pd.set_option('display.max_rows', 1000)
```

```python
df = pd.read_csv('hdas.csv')
#df[df['Depth'] == 3]
#df.describe()
df.shape
#df.head()
```

```python
depth_counts = df['Depth'].value_counts().sort_index()

# Plot the counts using a bar chart
depth_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Nodes at Each Depth Level')
plt.xlabel('Depth Level')
plt.ylabel('Number of Nodes')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python

hda_counts = Counter(df['HDA'])
top_hda = hda_counts.most_common(10)
hda_names, counts = zip(*top_hda)
plt.figure(figsize=(12, 8))
sns.barplot(list(hda_names), list(counts), palette="inferno_r")
plt.xticks(rotation=45)
plt.ylabel('Number of HDAs Instances', fontsize=12)
plt.xlabel('HDA Names')
plt.title('Top 10 HDA Instances in Scene', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

```
